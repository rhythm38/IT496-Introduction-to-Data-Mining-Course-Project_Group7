# We have scrapped data of  batting and bowling of world cup 2023 from ESPN Cricinfo,we have also cleaned/removed unwanted rows like injured players from the Dataset

```python
url="https://www.espncricinfo.com/records/tournament/bowling-most-wickets-career/icc-cricket-world-cup-2023-24-15338"
all_tables = pd.read_html(url)
all_tables[0]
```

```python
#Removing Injured players in 2023
df= all_tables[0]
rows_to_remove = [18,33,44,47,55,58,60,64,67,68,69,70,71,73,74,76,78]

# Remove rows by index
df_filtered = df.drop(rows_to_remove)

# Display the DataFrame after removing rows
print("\nDataFrame after removing rows:")
print(df_filtered)
```

```python
df_filtered.to_csv("Bowlers2023.csv")
df= pd.read_csv("Bowlers2023.csv")
df.head()
```
